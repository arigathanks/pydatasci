```
#common imports for this section
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
%matplotlib inline
```

```
#working with uniform distribution
```

```
#creating a list of random data from a uniform distribution
#uniform(lower bound, upper bound, size =(rows,columns))
uniform_data = random.uniform(0, 50, size=(5, 5))
```

```
#type of output
type(uniform_data)
```

```
#viewing the data array
uniform_data
```

```
sns.distplot(uniform_data,kde=False)
plt.show()
```

```
sns.distplot(random.uniform(size=100),kde=False)
plt.show()
```

```
sns.distplot(random.uniform(size=100),kde=True)
plt.show()
```

```
sns.distplot(random.uniform(size=100),kde=True, hist=False)
plt.show()
```

```
#working with Binomial distribution
```

```
#creating a list of random data from a uniform distribution
#binomial(number of trials, probability of success, size = N {number of scenarios executed})
binomial_data = random.binomial(n=10, p=0.75, size=20)
```

```
type(binomial_data)
```

```
binomial_data
```

```
sns.distplot(binomial_data,kde=False)
plt.show()
```

```
sns.distplot(random.binomial(n=10, p=0.25, size=1000),kde=False)
plt.show()
```

```
sns.distplot(random.binomial(n=10, p=0.25, size=1000),kde=True)
plt.show()
```

```
sns.distplot(random.binomial(n=10, p=0.25, size=1000),hist=False)
plt.show()
```

```
#working with Poisson distribution
```

```
#creating a list of random data from a uniform distribution
#poisson(lambda, size = N {number of scenarios executed})
poisson_data = random.poisson(lam = 5, size = 10)
```

```
#type of output
type(poisson_data)
```

```
#viewing the data array
poisson_data
```

```
sns.distplot(poisson_data,kde=False)
plt.show()
```

```
sns.distplot(random.poisson(lam = 10, size=100),kde=False)
plt.show()
```

```
sns.distplot(random.poisson(lam = 10, size=100),kde=False)
plt.show()
```

```
sns.distplot(random.poisson(lam = 10, size=100),kde=True, hist=False)
plt.show()
```

```
#working with Normal distribution
```

```
#creating a list of random data from a uniform distribution
#uniform(loc {mean}, scale {stdev}, size =(rows,columns))
normal_data = random.normal(loc = 50, scale = 15, size=(5, 5))
```

```
#type of output
type(normal_data)
```

```
#viewing the data array
normal_data
```

```
sns.distplot(normal_data,kde=False)
plt.show()
```

```
sns.distplot(random.normal(loc = 10, scale = 2, size=100),kde=False)
plt.show()
```

```
sns.distplot(random.normal(loc = 10, scale = 2, size=100),kde=False)
plt.show()
```

```
sns.distplot(random.normal(loc = 10, scale = 2, size=100),kde=True, hist=False)
plt.show()
```
