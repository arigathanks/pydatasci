```
#BEGIN ADVANCED PLOTTING
```

```
df_tips = sns.load_dataset("tips")
```

```
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df_tips)
plt.show()
```

```
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df_tips,
           markers=["o", "x"], palette = ["red","green"])
plt.show()
```

```
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df_tips)
plt.show()
```

```
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=df_tips)
plt.show()
```

```
sns.jointplot(x = "total_bill", y ="tip", data = df_tips, kind = "scatter")
plt.show()
```

```
sns.jointplot(x = "total_bill", y ="tip", data = df_tips, kind = "reg")
plt.show()
```

```
sns.jointplot(x = "total_bill", y ="tip", data = df_tips, kind = "kde")
plt.show()
```

```
import scipy.stats as stats
```

```
sns.jointplot(x = "total_bill", y ="tip", data = df_tips, kind = "reg").annotate(stats.pearsonr)
plt.show()
```
