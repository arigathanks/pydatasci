```
#common imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline

sns.set(color_codes = True)
```

```
df_titanic = sns.load_dataset("titanic")
df_iris = sns.load_dataset("iris")
df_tips = sns.load_dataset("tips")
```

```
sns.distplot(df_iris["sepal_length"], hist = True, kde = True, bins = 20, color = "blue", label = "Sepal Length")
sns.distplot(df_iris["sepal_width"], hist = True, kde = True, bins = 20, color = "red", label = "Sepal Width")
plt.legend()
plt.xlabel("Measure in CM")
plt.ylabel("Frequency")
plt.title("Sepal Characteristics", size = 15)
plt.show()
```

```
sns.lmplot(x = "total_bill", y = "tip", data = df_tips, hue = "sex",
           markers=["o", "x"], palette = ["red","black"])
plt.show()
```

```
sns.lmplot(x = "total_bill", y = "tip", data = df_tips, hue = "sex", col = "smoker",
           markers=["o", "x"], palette = ["red","black"])
plt.show()
```

```
sns.lmplot(x = "total_bill", y = "tip", data = df_tips, hue = "sex", col = "smoker", row = "time",
           markers=["o", "x"], palette = ["red","black"])
plt.show()
```

```
sns.jointplot(x = "total_bill", y = "tip", data = df_tips, kind = "scatter")
plt.show()
```
